```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Trendgenerator:

    def __init__(self, x_axis, y_axis, y_start) -> None:
        self.x_axis = x_axis
        self.y_axis = y_axis
        self.y_start = y_start
        

    def xyaxis(self):
        z = []
        
        for i in range(self.y_axis):
            x = random.randint(0,9)
            self.y_start = random.randint((self.y_start-x),(self.y_start+x))
            z.append(self.y_start)
        

        a = np.arange(self.x_axis)
        
        %matplotlib inline
        plt.xlabel("x_axis")
        plt.ylabel("y_axis")
        plt.plot(a, z, color='blue')



trend = Trendgenerator(3000, 3000, 60)
trend.xyaxis()
```
